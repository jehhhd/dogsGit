```python
import pandas as pd
import numpy as np
```

```python
sb = pd.read_csv('betwatch_prices.csv')
sb['Track']=sb['track_name'].str.lower()
sb=sb.rename(columns={'race_date':'Date'})
sb['DogName']=sb['name'].str.lower()

for i in range(1,11):
    sb['DogName'] = sb['DogName'].str.strip(str(i)+'.').str.strip()
    
sb=sb[['Date','DogName','price']]
sb
```

```python
a = pd.read_csv('allRaces1.csv')
a
```

```python
import os 
tracks = os.listdir('class2')
for track in tracks:
    if track != '.DS_Store': 
        print(track)
        data = pd.read_csv('class2/'+track)
        data2 = pd.merge(data,a,how='inner',on=['RaceId','RunTime'])
        for col in list(data2.columns):
            if('_x' in col):
                data2=data2.rename(columns={col:col.strip('_x')})

        cols = ['DogName','RunTime', 'Split', 'Place', 'Date', 'RaceId', 'Distance',
           'Weight', 'StartPrice', 'mean_place',
           'mean_in_race_pos', 'time_z_avg', 'split_z_avg', 'pred_split1', 't1',
           't2', 't3', 't4', 't5', 'last_5_avg', 'in_form', 'ctime_z_avg',
           'csplit_z_avg', 'pred_split2', 'at_cas_dist_time',
           'at_cas_dist_time_avg', 'at_cas_perc', 'in_form2', 'faster_inside',
           'next_to_faster', 'collision', 'first_out', 'pir1', 'pir2', 'pir3',
           'Win', 'b1', 'b2', 'b3', 'b4', 'b5', 'b6', 'b7', 'b8', 'i1', 'i2']
        data2=data2[cols]

        data3=pd.merge(data2,sb,how='left')
        data3.to_csv('class4/'+track)
```
