```python
df = pd.read_csv('betwatch_prices.csv')
df = df.loc[(df.price!=0)]
df
```

```python
ls = []
for x in list(df['name']):
    a = str(x)
    a = a.lstrip("1. ")
    a = a.lstrip("2. ")
    a = a.lstrip("3. ")
    a = a.lstrip("4. ")
    a = a.lstrip("5. ")
    a = a.lstrip("6. ")
    a = a.lstrip("7. ")
    a = a.lstrip("8. ")
    a = a.lstrip("9. ")
    a = a.lstrip("10. ")
    ls.append(a.lower())
    
df['name'] = ls
df
```

```python
bf = pd.read_csv('bsp_allRaces.csv')
bf['name'] = list(bf['DogName'])
bf['race_date'] = list(bf['date'])

df = pd.merge(df,bf, on=['name','race_date'],how='left').dropna()
df
```

```python
for x in list(set(df['Track'])):
    df1 = df.copy()
    df1 = df1[df1['Track']==x]
    
    
    ls1 = df1.loc[(df.price<2.1) & (df.price>1.2)]['price']
    ls2 = df1.loc[(df.price<2.1) & (df.price>1.2)]['StartPrice']
    
    ls3 = df1.loc[(df.price<3) & (df.price>2.1)]['price']
    ls4 = df1.loc[(df.price<3) & (df.price>2.1)]['StartPrice']
    
    ls5 = df1.loc[(df.price<5) & (df.price>3)]['price']
    ls6 = df1.loc[(df.price<5) & (df.price>3)]['StartPrice']

    diffs1 = ls1-ls2
    diffs2 = ls3-ls4
    diffs3 = ls5-ls6
    
    
    print(x)
    print('mean1: '+str(np.mean(diffs1)))
    print('sd1: '+str(np.std(diffs1)))
    print(' ')
    print('mean2: '+str(np.mean(diffs2)))
    print('sd2: '+str(np.std(diffs2)))
    print(' ')
    print('mean3: '+str(np.mean(diffs3)))
    print('sd3: '+str(np.std(diffs3)))
    print('----------------')
    
```
