```python
import pandas as pd
import numpy as np
import os
```

```python
files = os.listdir('FastTrack')
files.sort()
dog_files=files[1:int(len(files)/2)+1]
race_files=files[1+int(len(files)/2):]
pd.DataFrame([dog_files,race_files]).T
```

```python
df = ''
filepath = 'FastTrack/'
for i in range(len(dog_files)):
    
    dogs = pd.read_csv(filepath + dog_files[i])
    ids = pd.read_csv(filepath + race_files[i])
#     print(ids.columns)
    dates = pd.to_datetime(ids['date'], format='%d %b %y')
    ids['date']=dates
    ids['Distance']=ids['Distance'].str[:-1].astype(int)
    ids=ids.drop(columns=['RaceNum','RaceName','RaceTime'])
    ids=ids.rename(columns={'@id':'RaceId'})
    merged = pd.merge(dogs,ids,on='RaceId',how='inner')
    merged['PIR']=merged['PIR'].astype(str)
    if(i==0):
        df=merged
    else:
        df = pd.concat([df,merged],ignore_index=True)
    print(i)
df = df.drop(columns=['Handicap','Margin1','Margin2','Prizemoney','Checks','Comments','Rug','@id','TrainerId'])

df
```

```python
df2 = df.copy()
# df2 = df2.drop(columns=['Handicap','Margin1','Margin2','Prizemoney','Checks','Comments','RaceId','Rug','@id','TrainerId'])
df2['DogName'] = df2['DogName'].str.lower()
df2['StartPrice'] = df2['StartPrice'].str.replace(r'[^0-9.]', '', regex=True).astype(float)
df2['Place']=df2['Place'].str.replace(r'[^0-9.]', '', regex=True)
df2=df2[df2['Place']!='']
df2=df2.rename(columns={'SplitMargin':'Split'})
df2['Track']=df2['Track'].str.lower()
df2['TrainerLName']=df2['TrainerName'].str.split(' ').str[1]
df2['TrainerFName']=df2['TrainerName'].str.split(' ').str[0]
df2 = df2.drop(columns=['TrainerName'])
df2['PIR']=np.where(df2['PIR'].str.contains('/'),df2['PIR'].str.split('/').str[1],df2['PIR'])
df2
```

```python
scrape = ''
for i in range(1,98):
    file = 'scrapes/scrape'+str(i)+'.csv'
    x = pd.read_csv(file)
#     print(x.shape[0])
    if(i==1):
        scrape=x
    else:
        scrape=pd.concat([scrape,x],ignore_index=True)
scrape
```

```python
df3 = scrape.copy()
df3 = df3.drop(columns=['Mgn'])
for col in list(df3.columns):
    if('Unnamed' in col):
        df3 = df3.drop(columns=[col])
        
        
df3 = df3.rename(columns={'Plc':'Place','Name (Box)':'DogName','Wgt':'Weight','In Run':'PIR','SP':'StartPrice','Trainer':'TrainerName','Time':'RunTime'})
df3=df3[df3['DogName'].str[-2]!='(']
df3['Box']=df3['DogName'].str[-2].astype(int)
df3['DogName']=df3['DogName'].str[:-4].str.lower()
df3['Weight']=df3['Weight'].str.strip('kg')
df3['StartPrice']=df3['StartPrice'].str[1:]
df3['TrainerLName']=df3['TrainerName'].str.split(' ').str[-1]
df3['TrainerFName']=df3['TrainerName'].str.split(' ').str[0].str[0]
df3 = df3.drop(columns='TrainerName')
df3['Track']=df3['Track'].apply(lambda x: x.replace("-", " "))
df3
```
