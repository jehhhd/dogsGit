```python
import os
import pandas as pd
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('parameters.json','r') as json_file:
    unique_track_dist = json.load(json_file)
    
```

```python
files = os.listdir('class')
files.remove('warr_test.csv')
files.remove('.DS_Store')
files.remove('warrnambool1.csv')

for f in files[:]:
    file = 'class/'+f
#     print(f)
    track = f.split('.')[0]
    
    print(f,track=='meadows')
    
    if(track=='meadows'):
        print('here')
        track='the meadows'
    params_all = unique_track_dist[track]
    
    df = pd.read_csv(file)
    dists = df['Distance'].unique()
    
    
    
    for d in dists:
        params = params_all[str(d)]
        df['at_cas_dist_time'] = np.where( df['Distance']==d,(df['at_cas_dist_time']-params[2])/params[3],df['at_cas_dist_time'])
        df['at_cas_dist_time_avg'] = np.where( df['Distance']==d,(df['at_cas_dist_time_avg']-params[2])/params[3],df['at_cas_dist_time_avg'])
        
    df=df[abs(df['at_cas_dist_time'])<3.5]
    df=df[abs(df['at_cas_dist_time_avg'])<3.5]
    
    df.to_csv('class2/'+f,index=False)
    
    
        
        
    
    
    print(df['at_cas_dist_time'].describe())
```
