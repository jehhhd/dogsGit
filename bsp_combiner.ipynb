```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
files = os.listdir('bsp')
files = [x for x in files if ('.D' not in x) and ('checkpoints' not in x)]
# files=files[-1:]
files.sort()
print(files)
for file in files:
    if(file==files[0]):
        print(True)
        df = pd.read_csv('bsp/'+file)
        df['date']=pd.to_datetime(df['date']).astype(str)
        
    else:
        df_temp = pd.read_csv('bsp/'+file)
        df_temp['date']=pd.to_datetime(df_temp['date']).astype(str)
        df = pd.concat([df,df_temp])


df = df.rename(columns={'track':'Track','name':'DogName'})

df['Track']=df['Track'].str.strip()
df['DogName']=df['DogName'].str.strip()
df['date']=df['date'].str.strip()


df['menu_day'] = df['MENU_HINT'].str.extract(r'(\d{1,2})\D+(\w+)')[0].astype(str)
df['menu_month']=df['MENU_HINT'].str.split(' ').str[-1].str[:3]
df['menu_yr']=df['date'].str[:4]

df['date']=df['menu_day']+'-'+df['menu_month']+'-'+df['menu_yr'].astype(str)
df['date']=pd.to_datetime(df['date'],format="%d-%b-%Y", errors='coerce').astype(str)
df = df.dropna(subset=['date'])
df = df.drop(['menu_day', 'menu_month', 'menu_yr'], axis=1)
df
```

```python
trks = ['richmond straight','shepparton', 'maitland', 'ballarat', 'sandown park', 'goulburn', 'hobart', 'dapto', 'angle park', 'warrnambool', 'warragul', 'albion park', 'northam', 'traralgon', 'healesville', 'richmond', 'ipswich', 'geelong', 'casino', 'the gardens', 'cannington', 'mount gambier', 'dubbo', 'bendigo', 'bulli', 'horsham', 'wentworth park', 'the meadows', 'sale', 'darwin', 'gawler', 'grafton', 'nowra', 'bathurst', 'launceston', 'townsville', 'devonport', 'lismore', 'rockhampton', 'cranbourne', 'strathalbyn', 'wagga', 'cairns', 'mandurah', 'temora', 'gosford', 'capalaba', 'murray bridge', 'angle park', 'ipswich', 'bundaberg', 'gunnedah', 'taree']
races = pd.read_csv('allRaces1.csv')
races = races[races['Track'].isin(trks)]

char = "'"
races['DogName']=races['DogName'].str.replace(char,'').str.replace('.','')

races['Track']=races['Track'].str.strip()
races['DogName']=races['DogName'].astype(str).str.strip()
races['date']=races['date'].astype(str).str.strip()



races2 = pd.merge(races,df,on=['DogName','date','Track'],how='left')
races2 = races2.dropna(subset=['Place'])
races2
```
