```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pandas as pd
from selenium.webdriver.common.by import By
from datetime import datetime
from selenium.common.exceptions import NoSuchElementException
import time 
```

```python
months = [1]
days=[31]
dates=[]
for i in range(len(months)):
    string = '2024-' +str(months[i])
    for j in range(31,days[i]+1):
        if(len(str(j))==1):
            string2 = string +'-0'+str(j)
        else:
            string2 = string+'-'+str(j)
        print(string2)
        dates.append(string2)
```

```python
dates
```

```python
base='https://www.thegreyhoundrecorder.com.au/results/search/'
driver = webdriver.Safari()
links=[]
for i in dates:
    url = base + i +'/'
    driver.get(url)
    time.sleep(2)
    rows = driver.find_elements(By.CLASS_NAME,'meeting-row')
    for row in rows:
        title = row.find_element(By.CLASS_NAME,'meeting-row__title').text
        
        state = title.split('(')[1].strip(')')
        link=row.find_element(By.TAG_NAME,'a').get_attribute('href')
        print(title,state,link)
        if(state not in ['NZL','GB']):
            
            
            if(link not in links):
                links.append(link)
    print(i)
            
links
```

```python
links1 = links.copy()
nz = ['addington','ascot-park','cambridge','hatrick','manawatu','manukau','sunderland','pelaw-grange','youghal','tralee','thurles-park','shelbourne-park','limerick','galway','dundalk','drumbo-park','kilkenny','curraheen-park']
uk = ['doncaster','harlow','kinsley','monmore','nottingham','perry-barr','romford','sheffield','suffolk-downs','swindon','towcester','yarmouth','central-park','crayford','hove','newcastle','oxford','henlow','clonmel']

for x in links1:
    for i in nz:
        if i in x.split('/'):
            links.remove(x)
    for i in uk:
        if i in x.split('/'):
            links.remove(x)

links.insert(0,'https://www.thegreyhoundrecorder.com.au/results/pelaw-grange/226121/')
            
links
```

```python
import warnings
warnings.simplefilter("ignore", category=FutureWarning)
mega = ''
driver = webdriver.Chrome()
for j in range(0,len(links)):
    url=links[j]
    driver.get(url)

    races = len(driver.find_elements(By.CLASS_NAME,'meeting-events-nav__item'))
    date = driver.find_element(By.CLASS_NAME,'results-meeting-events__date').text
    date = ' '.join(date.split(' ')[1:])
    date_format = '%B %d, %Y'
    date = datetime.strptime(date, date_format)
    date = date.strftime('%d/%m/%Y')
    track = url.split('/')[-3]

#     for i in range(1,2):
    for i in range(1,races+1):

        URL=url+str(i)+'/'
#         driver = webdriver.Chrome()
        driver.get(URL)
        try:
            details = driver.find_element(By.CLASS_NAME,'results-event')
            dist = details.find_element(By.CLASS_NAME,'meeting-event__header-distance').text[:-1]
            grade = details.find_element(By.CLASS_NAME,'meeting-event__header-class').text


            tab = driver.find_element(By.CLASS_NAME,'results-event__table')
            df= pd.read_html(tab.get_attribute('outerHTML'))[0]
            df=df.drop(columns=['Rug','Sire','Dam'])

            df=df[df['Plc']!='SCR']
            df['Distance']=int(dist)
            df['RaceGrade']=grade
            df['date']=date
            df['Track']=track

            if(j==0):
                mega = df
            else:
                mega = pd.concat([mega,df],ignore_index=True)
        except NoSuchElementException:
            pass
        
    print(j)
        
    

```

```python
mega.to_csv('scrapes/scrape100.csv')
```
